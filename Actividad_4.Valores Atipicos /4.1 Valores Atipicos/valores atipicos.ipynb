```python
#Creo 2 dataframe  para podeer procesar los outliers
#cuantitativas= data.iloc[ :, 7:14]
#cualitativas= data.iloc[ :, [0,1]]

cualitativas = data.select_dtypes(include=['object', 'datetime64[ns]'])
cuantitativas= data.select_dtypes(include=['int64', 'float64'])
```

```python
#realizamos diagrama de cja o  bigote de cada columna del datframe

fig = plt.figure(figsize= (15, 8))
cuantitativas.plot(kind= "box", vert=False) #respeta las mayusculas
plt.title("Valores Atipicos del dataframe")
plt.show() #dibujamos el dataframe
```

```python
cuantitativas.info()
```

```python
#METODO DE APLICAIONES DE DESVIACION ESTANDAR . ENCUENTRO LOS VALORES EXTREMOS
y=cuantitativas
Limite_Superior= y.mean()+3*y.std()
Limite_Inferior= y.mean()-3*y.std()
print("Limite Superior permitido", Limite_Superior)
print("LImite inferior permitido", Limite_Inferior)
```

```python
#obtener datos y los outliers se conviertenen nulos en le dtaframe
data3= cuantitativas [(y<=Limite_Superior)&(y>=Limite_Inferior)]
data3
```

```python
#corroboramos valores nulos del dataframe
valores_nulos = data3.isnull().sum()
valores_nulos
```

```python
#reemplazamos valores atipicos (nulos) del dtaaframe con mean
#realizamos copia del datfarame
data_clean=data3.copy()
data_clean=data_clean.fillna(round(data3.mean(),1))
data_clean
```

```python
valores_nulos=data_clean.isnull().sum()
valores_nulos
```

```python
#Metodo aplicando cuartiles encuemtra cuartiles 0.25 y 0.75
y= cuantitativas

percentile25= y. quantile(0.25)
percentile75= y.quantile(0.75)
iqr=percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superiot permitido", Limite_Superior_iqr)
print("Limite_Inferior permitido", Limite_Inferior_iqr)
```

```python
data3_iqr= cuantitativas [(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data3_iqr
```

```python
valores_nulos=data3_iqr.isnull().sum()
valores_nulos
```

```python
#reemplazamos valores atipicos (nulos) del dataframe con mean
#realizamos una copia del data frame

data4_iqr=data3_iqr.copy()
data4_iqr=data4_iqr.fillna(round(data3_iqr.mean(),1))
data4_iqr
```

```python
#unimos el informe de cuantitativo limpio co el dataframe cualitativo

Datos_limpios= pd.concat([cualitativas, data4_iqr], axis=1)
Datos_limpios
```
